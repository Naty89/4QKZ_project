```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

fpocket_file = "4QKZ_info.txt"

data = []  # store all pocket info here
current_pocket = {}

with open(fpocket_file) as f:
    for line in f:
        line = line.strip()
        if line.startswith("Pocket"):
            if current_pocket:
                data.append(current_pocket)
            current_pocket = {"Pocket": int(line.split()[1])}
        elif "Volume" in line:
            current_pocket["Volume"] = float(line.split(":")[1].strip())
        elif "Hydrophobicity score" in line:
            current_pocket["Hydrophobicity"] = float(line.split(":")[1].strip())
        elif "Polarity score" in line:
            current_pocket["Polarity"] = int(line.split(":")[1].strip())

# append the last pocket
if current_pocket:
    data.append(current_pocket)

# converting and sorting DataFrame
df = pd.DataFrame(data)
df = df.sort_values("Pocket")

#pocket characterization

fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 cols

#volumes
axes[0].bar(df["Pocket"], df["Volume"], color="skyblue")
axes[0].set_xlabel("Pocket")
axes[0].set_ylabel("Volume (Å³)")
axes[0].set_title("Pocket Volumes")
axes[0].grid(alpha=0.3)

#polarity
axes[1].bar(df["Pocket"], df["Polarity"], color="lightgreen")
axes[1].set_xlabel("Pocket")
axes[1].set_ylabel("Polarity Score")
axes[1].set_title("Pocket Polarity")
axes[1].grid(alpha=0.3)

#hydrophobicity
axes[2].bar(df["Pocket"], df["Hydrophobicity"], color="salmon")
axes[2].set_xlabel("Pocket")
axes[2].set_ylabel("Hydrophobicity Score")
axes[2].set_title("Pocket Hydrophobicity")
axes[2].grid(alpha=0.3)

plt.show()
```

```python
import py3Dmol

# === Step 1: Load the main output pdb from fpocket ===
pdb_file = "4QKZ_out.pdb"

with open(pdb_file, "r") as f:
    pdb_data = f.read()

# === Step 2: View it in py3Dmol ===
view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, "pdb")

# === Step 3: Style the protein and pockets ===
# Protein backbone in white cartoon
view.setStyle({"chain": "A"}, {"cartoon": {"color": "white"}})

# Pocket spheres colored differently (fpocket marks them as HETATM)
colors = [
    "red", "orange", "yellow", "green", "cyan", "blue", "purple", "magenta", "pink", "lime"
]

# Color each pocket’s alpha spheres differently
for i, color in enumerate(colors):
    view.addStyle({"resn": f"PK{i+1}"}, {"sphere": {"color": color, "radius": 1.0}})

view.zoomTo()
view.show()
```

```python
import py3Dmol
import glob

protein_file = "4QKZ.pdb"
pockets_dir = "pockets"

# Load protein
with open(protein_file) as f:
    pdb_data = f.read()

view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, "pdb")

# Protein backbone + semi-transparent surface
view.setStyle({}, {
    "cartoon": {"color": "white"},
    "surface": {"opacity": 0.3, "color": "white"}  # lower opacity so pockets show through
})

# Automatically get all pockets
pockets = sorted(glob.glob(f"{pockets_dir}/pocket*_atm.pdb"))

# Cycle colors for pockets
colors = ["red", "blue", "green", "orange", "magenta", "cyan", "yellow", "purple"]

for i, pocket_file in enumerate(pockets):
    with open(pocket_file) as f:
        pocket_atoms = f.read()
    
    # Add pocket as a **surface** instead of spheres
    view.addModel(pocket_atoms, "pdb")
    view.setStyle(
        {"model": i + 1},
        {"surface": {"opacity": 0.7, "color": colors[i % len(colors)]}}
    )

view.zoomTo()
view.show()
```
